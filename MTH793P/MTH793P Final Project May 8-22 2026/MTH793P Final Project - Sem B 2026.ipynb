### Effect of the RBF bandwidth parameter (gamma)

The RBF affinity kernel is $W_{ij} = \exp(-\gamma \|x_i - x_j\|^2)$.  
A **high gamma** means only pixels with nearly identical colours are strongly connected, fragmenting the affinity graph into many small local neighbourhoods. Below we test `gamma = 200` (vs the baseline `gamma = 50`) to observe this effect.

````python
# High gamma tightens the affinity kernel. Observe the effect vs baseline gamma=50
spectral_3_ga_200 = SpectralClustering(n_clusters=3, affinity='rbf', gamma=200.0, random_state=0, n_init=5)
_ = plot_clusters(spectral_3_ga_200)
````

### Discussion

**Effect of `n_clusters`:**  
With `n_clusters = 2` the image is split into two broad regions (roughly light vs dark), losing meaningful distinctions (underfit). At `n_clusters = 3` (the baseline) three visually coherent zones emerge - this aligns well with the various land types. Increasing to 4 or 5 reveals finer sub-regions (e.g. distinguishing different vegetation shades), which still look geographically plausible. Reconstruct the **first image** in `X_sample` using k = 2, 5, 20, 50, 100 eigendigits.
For each k, compute the reconstruction as:
$$\hat{x} = \text{mean\_digit} + (x_{\text{centered}} \cdot V_k^\top) \cdot V_k$$
where $V_k$ contains the first k rows of **Vt**.

Display the original and the 5 reconstructions in a single row. Store the reconstruction
for k=50 in **<font color='red'>recon_50</font>**.

````python
# Get the first centered image
x_centered = X_centered[0]

# k values for reconstruction
k_values = [2, 5, 20, 50, 100]

# Compute reconstructions for each k
reconstructions = {}
for k in k_values:
    # Extract the first k eigendigits (rows of Vt)
    V_k = Vt[:k, :]
    
    # Project the centered image onto the eigendigits
    projection = x_centered @ V_k.T
    
    # Reconstruct: mean_digit + projection * V_k
    reconstruction = mean_digit + projection @ V_k
    reconstructions[k] = reconstruction

# Store the reconstruction for k=50
recon_50 = reconstructions[50]

# Display original and reconstructions in a single row
fig, axes = plt.subplots(1, 6, figsize=(15, 2.5))

# Display original image
axes[0].imshow(vec2img(X_sample[0]), cmap='gray')
axes[0].set_title('Original')
axes[0].axis('off')

# Display reconstructions for each k
for idx, k in enumerate(k_values):
    axes[idx + 1].imshow(vec2img(reconstructions[k]), cmap='gray')
    axes[idx + 1].set_title(f'k={k}')
    axes[idx + 1].axis('off')

plt.tight_layout()
plt.show()
````

## Section B — Robust PCA via IALM (20 marks)

\
We now implement **Robust PCA** to decompose a corrupted image $M = L + S$ into a
**low-rank component** $L$ and a **sparse corruption** $S$, by solving:
$$\min_{L,S}\; \|L\|_* + \lambda\|S\|_1 \quad\text{s.t.}\quad L+S=M.$$

We use the **Inexact Augmented Lagrangian Method (IALM)**:
$$\begin{aligned}
L^{(k+1)} &= \mathcal{D}_{1/\mu}(M - S^{(k)} + Y^{(k)}/\mu),\\
S^{(k+1)} &= \mathcal{T}_{\lambda/\mu}(M - L^{(k+1)} + Y^{(k)}/\mu),\\
Y^{(k+1)} &= Y^{(k)} + \mu(M - L^{(k+1)} - S^{(k+1)}),\\
\mu^{(k+1)} &= \min(\rho\,\mu^{(k)},\;\mu_{\max}),
\end{aligned}$$
where $\mathcal{D}_\tau$ soft-thresholds singular values and
$\mathcal{T}_\epsilon(x) = \text{sign}(x)\max(|x|-\epsilon,0)$ is element-wise soft-thresholding.

Stop when $\|M - L^{(k)} - S^{(k)}\|_F / \|M\|_F \le \text{tolerance}$.

The penalty parameter is initialized following Lin et al. (2010):
$$\mu_0 = 1.25 \;/\; \|M\|_2,$$
where $\|M\|_2$ is the spectral norm (largest singular value of $M$).
This ties the initial SVT threshold directly to the scale of the data,
ensuring the low-rank and sparse components are separated correctly from
the very first iteration.

**Step 1: Implement the two thresholding operators.**

- **<font color='red'>singular_value_threshold(A, tau)</font>** — compute SVD of A,
  soft-threshold singular values by tau, return reconstructed matrix.
- **<font color='red'>element_threshold(A, eps)</font>** — element-wise soft-threshold:
  $\text{sign}(A)\max(|A|-\varepsilon,\,0)$.

````python
def element_threshold(A, eps):
    A = np.array(A, dtype=float)
    return np.sign(A) * np.maximum(np.abs(A) - eps, 0.0)


def singular_value_threshold(A, tau):
    A = np.array(A, dtype=float)
    U, s, Vt = svd(A, full_matrices=False)
    s_thr = np.maximum(s - tau, 0.0)
    return (U * s_thr) @ Vt
````

````python
### DO NOT REMOVE/EDIT THIS CELL ###

# element_threshold
assert_almost_equal(element_threshold(np.array([[0.3,-0.2],[1.0,-1.5]]),0.5),
                    np.array([[0.0,0.0],[0.5,-1.0]]))
# singular_value_threshold: rank should not increase
B = np.outer([1.0,2.0,3.0],[1.0,0.5])  # rank-1
D = singular_value_threshold(B, 0.1)
assert D.shape == B.shape
# large tau => zero matrix
assert_almost_equal(singular_value_threshold(np.eye(3)*0.1, 1.0), np.zeros((3,3)))
print('--- TEST PASSED ---')
````

**Step 2: Implement the IALM Robust PCA solver.**

Write **<font color='red'>robust_pca(M, lam, mu, rho, mu_max, tolerance, max_iter, iter_print)</font>**
with defaults:
- `lam = None` → $1/\sqrt{\max(D,N)}$
- `mu = None`  → $1.25 \;/\; \|M\|_2$ &nbsp; (use `svd(M, compute_uv=False)[0]` for $\|M\|_2$)
- `mu_max = None` → $10^7 \times \mu_0$
- `rho = 1.5`, `tolerance = 1e-5`, `max_iter = 500`, `iter_print = 50`

Return **(L, S)**.

````python
def robust_pca(M, lam=None, mu=None, rho=1.5, mu_max=None,
               tolerance=1e-5, max_iter=500, iter_print=50):
    M = np.array(M, dtype=float)
    D, N = M.shape

    if lam is None:
        lam = 1.0 / np.sqrt(max(D, N))
    if mu is None:
        mu = 1.25 / svd(M, compute_uv=False)[0]
    if mu_max is None:
        mu_max = 1e7 * mu

    L = np.zeros_like(M)
    S = np.zeros_like(M)
    Y = M / max(norm(M, 'fro'), np.finfo(float).eps)
    norm_M = norm(M, 'fro')

    for k in range(max_iter):
        L = singular_value_threshold(M - S + Y / mu, 1.0 / mu)
        S = element_threshold(M - L + Y / mu, lam / mu)
        residual = M - L - S
        residual_norm = norm(residual, 'fro')
        if residual_norm / norm_M <= tolerance:
            break
        Y = Y + mu * residual
        mu = min(rho * mu, mu_max)

    return L, S
````

The cell below tests your robust_pca implementation on a small synthetic matrix of size 20 × 25. It constructs a rank-2 ground-truth low-rank component L and a sparse corruption S with 10% non-zero entries set to ±5, then checks that your decomposition recovers L+S=M with relative error below 1%. Complete the code. 

````python
### DO NOT REMOVE/EDIT THIS CELL ###

np.random.seed(123)
D_r, N_r = 20, 25
U_r = np.random.randn(D_r, 2); V_r = np.random.randn(2, N_r)
L_true = U_r @ V_r
S_true = np.zeros((D_r, N_r))
idx_c = np.random.choice(D_r*N_r, size=int(0.10*D_r*N_r), replace=False)
S_true.flat[idx_c] = np.random.choice([-5.0, 5.0], size=len(idx_c))
M_t = L_true + S_true
L_r, S_r = robust_pca(M_t)

rel_err = norm(M_t - L_r - S_r, 'fro') / norm(M_t, 'fro')
assert rel_err < 0.01, f'Error too large: {rel_err:.4f}'
print(f'Relative error: {rel_err:.6f}')
print('--- TEST PASSED ---')
````

1. The low-rank component $L$ is the recovered structured image. It is the result of projecting the corrupted image onto a low-dimensional structure/manifold where the "true" image lives. Component S represents the left over anomaly $S = X - L$. These are the pixels that do not fit the coherent structure of the image (localised anomolies, corruptions).

2. Standard PCA is extremely brittle in the presence of large outliers. This is because standard PCA looks for the best rank-$k$ estimate of the low-rank $L$ via:
    $$
    \begin{aligned}
    \text{minimize} \quad & \|X-L\|_F \\
    \text{subject to} \quad & \operatorname{rank}(L) \le k
    \end{aligned}
    $$

    Where it's clear that large anomalies dominate.

    Robust PCA considers a different optimisation problem:
    $$
    \begin{aligned}
    \min_{L,S} \quad & \|L\|_* + \lambda \|S\|_1 \\
    \text{subject to} \quad & X = L + S
    \end{aligned}
    $$

    Separating the data into a low-rank and sparse components. Which performs much better with outliers since a single low-rank model isn't forced to explain large anomalous data. 

3. Setting $\lambda$ larger leads to more noise remaining in low-rank $L$, hence resembling less the original image. The sparse image $S$ also becomes even sparser, keeping only the strongest anomalies.

    Setting $\lambda$ **smaller** leads to more of the true underlying image being mistaken for noise, and so moved into $S$.

# Part III - Autoencoder for Dimensionality Reduction (30 marks)

\
We implement a **shallow autoencoder** from scratch using only NumPy and apply it to
compress and reconstruct handwritten digit images.

The network has the structure:
$$z = \sigma(W_1 x + b_1), \qquad \hat{x} = W_2 z + b_2,$$
where $\sigma$ is the **sigmoid** activation applied element-wise and the output layer is
**linear**. Parameters: $W_1\in\mathbb{R}^{p\times d}$, $b_1\in\mathbb{R}^p$,
$W_2\in\mathbb{R}^{d\times p}$, $b_2\in\mathbb{R}^d$.

Loss: $\mathcal{L}=\frac{1}{n}\|{\hat{X}}-X\|_F^2$. Optimized via **mini-batch gradient descent**.

---

````python
f = open('digits.npy','rb')
digits_all = np.load(f); f.close()
N_train = 500
X_train = digits_all[:N_train].astype(float).reshape(N_train, -1) 
print('Training data shape:', X_train.shape)
````

**Step 1: Sigmoid and its derivative.**

- **<font color='red'>sigmoid(z)</font>** — element-wise $1/(1+e^{-z})$.
- **<font color='red'>sigmoid_deriv(z)</font>** — element-wise $\sigma(z)(1-\sigma(z))$.

````python
def sigmoid(z):
    return 1.0 / (1.0 + np.exp(-z))

def sigmoid_deriv(z):
    sig = sigmoid(z)
    return sig * (1.0 - sig)
````

\
**Step 2: Forward pass.**

**<font color='red'>forward(X, W1, b1, W2, b2)</font>** — takes input batch `(n, d)` and returns
**(z, x_hat)**:
- $z = \sigma(X W_1^\top + b_1)$, shape `(n, p)`.
- $\hat{x} = z W_2^\top + b_2$, shape `(n, d)`.

````python
def forward(X, W1, b1, W2, b2):
    # (n, p) matrix of all encoded samples.
    z = sigmoid(X @ W1.T + b1)

    # (n, d) matrix of all reconstructed samples.
    x_hat = z @ W2.T + b2
    
    return z, x_hat
````

\
**Step 3: Backward pass.**

**<font color='red'>backward(X, z, x_hat, W1, b1, W2, b2)</font>** — returns gradients
**(dW1, db1, dW2, db2)** of $\mathcal{L}=\frac{1}{n}\|\hat{X}-X\|_F^2$.

*Hint: start from the output layer error $\delta_2 = \frac{2}{n}(\hat{X}-X)$, then
backpropagate through $W_2$, through the sigmoid, and through $W_1$.*

````python
def backward(X, z, x_hat, W1, b1, W2, b2):
    n = X.shape[0]
    
    # I keep track of matrix/vector shapes to help make sense of the overall process
    # Output layer error: dL/d(x_hat)
    delta_2 = (2.0 / n) * (x_hat - X)  # shape (n, d)
    
    # Gradients for W2 and b2
    # x_hat = z @ W2.T  =>  dL/dW2 = (dL/d(x_hat)).T @ z
    dW2 = delta_2.T @ z              # shape (d, p)
    db2 = np.sum(delta_2, axis=0)   # shape (d,)
    
    # Backpropagate to hidden layer
    # x_hat = z @ W2.T  =>  dL/dz = dL/d(x_hat) @ W2  (not W2.T)
    dz = delta_2 @ W2               # shape (n, p)
    delta_1 = dz * z * (1 - z)     # sigmoid derivative: sigma*(1-sigma)
    
    # Gradients for W1 and b1
    # z = sigmoid(X @ W1.T + b1)  =>  dL/dW1 = delta_1.T @ X
    dW1 = delta_1.T @ X             # shape (p, d)
    db1 = np.sum(delta_1, axis=0)  # shape (p,)
    
    return dW1, db1, dW2, db2
````

**Step 4: Training loop.**

**<font color='red'>train_autoencoder(X, p, lr, n_epochs, batch_size, random_state)</font>**:
- `lr=0.01`, `n_epochs=50`, `batch_size=128`, `random_state=0`
- Initialize weights: Gaussian std=0.01, biases=0.
- Return **(W1, b1, W2, b2, loss_history)**.

After training: plot the loss curve and display **5 original vs 5 reconstructed** digit images
side by side.

````python
def train_autoencoder(X, p, lr=0.01, n_epochs=50, batch_size=128, random_state=0):
    np.random.seed(random_state)
    
    n, d = X.shape
    
    # Initialise weights with Gaussian std=0.01, biases to zero
    W1 = np.random.randn(p, d) * 0.01
    b1 = np.zeros(p)
    W2 = np.random.randn(d, p) * 0.01
    b2 = np.zeros(d)
    
    loss_history = []
    
    for epoch in range(n_epochs):
        # Shuffle data each epoch
        indices = np.random.permutation(n)
        X_shuffled = X[indices]
        
        epoch_loss = 0.0
        
        # Mini-batch gradient descent
        for i in range(0, n, batch_size):
            X_batch = X_shuffled[i:i + batch_size]
            
            z, x_hat = forward(X_batch, W1, b1, W2, b2)
            
            epoch_loss += np.mean((x_hat - X_batch) ** 2) * X_batch.shape[0]
            
            dW1, db1, dW2, db2 = backward(X_batch, z, x_hat, W1, b1, W2, b2)
            
            W1 -= lr * dW1
            b1 -= lr * db1
            W2 -= lr * dW2
            b2 -= lr * db2
        
        avg_loss = epoch_loss / n
        loss_history.append(avg_loss)
        
        if (epoch + 1) % 10 == 0:
            print(f'Epoch {epoch+1}/{n_epochs}, Loss: {avg_loss:.6f}')
    
    # Plot loss curve
    plt.figure(figsize=(8, 5))
    plt.plot(loss_history, linewidth=2)
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.title(f'Autoencoder Training Loss ({n_epochs} epochs)')
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()
    
    # Display 5 original vs 5 reconstructed digits
    _, x_hat_test = forward(X[:5], W1, b1, W2, b2)
    fig, axes = plt.subplots(2, 5, figsize=(12, 5))
    for i in range(5):
        axes[0, i].imshow(X[i].reshape(24, 24), cmap='gray')
        axes[0, i].set_title(f'Original {i}')
        axes[0, i].axis('off')
        axes[1, i].imshow(x_hat_test[i].reshape(24, 24), cmap='gray')
        axes[1, i].set_title(f'Reconstructed {i}')
        axes[1, i].axis('off')
    plt.tight_layout()
    plt.show()
    
    return W1, b1, W2, b2, loss_history
````

## Comparison of Autoencoder with PCA

**1. Which method gives better visual reconstructions? Does this match your expectation?**

PCA achieves the lowest reconstruction MSE (see printed values above), followed by the 200-epoch autoencoder, with the 50-epoch autoencoder performing worst. This matches expectations. Handwritten digits lie close to a low-dimensional linear subspace, so PCA's globally optimal rank-32 projection is difficult to beat. The 50-epoch autoencoder has not converged fully (training loss is still decreasing at 50 epochs and the reconstructions are still very blurry. While the 200-epoch model closes some of that gap it still cannot match PCA's closed-form solution. The improvement from 50 to 200 epochs confirms the network is still learning, but not yet at a local minimum.

**2. What objective function difference explains the quality gap?**

All three methods minimise the same squared reconstruction error, but with very different approaches to optimisation. PCA finds the globally optimal rank-32 linear approximation (Eckart-Young Theorem) of the centred data in closed form via SVD - there is no local-minimum problem. The autoencoder minimises the same loss through mini-batch gradient descent on a nonlinear sigmoid network, which can get trapped in local minima and converges slowly with small initialisations. The 50-epoch model shows this gap most clearly; the 200-epoch model reduces it by allowing more gradient steps, but the optimisation disadvantage remains.

**3. A linear autoencoder (identity activation instead of sigmoid) is mathematically equivalent to which method, and why?**

A linear autoencoder with squared-error loss is mathematically equivalent to PCA. Because both the encoder and decoder are linear, their composition spans a linear subspace of dimension at most $p$. Minimising the reconstruction loss over all such networks is therefore the same problem as finding the best rank-$p$ approximation of the centred data - which is exactly what PCA solves via the SVD. Unlike the sigmoid autoencoder, the linear version has no local minima and converges to the principal subspace regardless of initialisation.
